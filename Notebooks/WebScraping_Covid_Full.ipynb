```python
!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
wd.get("https://www.worldometers.info/coronavirus/")
```

```python
countries = []
confirmeds = []
new_cases = []
deaths = []
new_deaths = []
recov = []
new_recov = []
act_cases = []
total_tests = []
population = []

# /html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[7]/td[2]/a

for i in range(5,223):
  try:
    country = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[2]/a').text
    countries.append(country)

    confirm = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[3]').text
    confirmeds.append(confirm)

    new_c = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[4]').text
    new_cases.append(new_c.replace('+', ''))

    dead = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[5]').text
    deaths.append(dead)

    new_d = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[6]').text
    new_deaths.append(new_d.replace('+', ''))

    rec = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[7]').text
    recov.append(rec)

    new_r = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[8]').text
    new_recov.append(new_r.replace('+', ''))

    act = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[9]').text
    act_cases.append(act)

    test = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[13]').text
    total_tests.append(test)

    pop = wd.find_element_by_xpath(f'/html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[{i}]/td[15]').text
    population.append(pop)

    # /html/body/div[3]/div[3]/div/div[6]/div[1]/div/table/tbody[1]/tr[87]/td[13]
  except Exception as e:
    print(e)
```

```python
import pandas as pd

df = pd.DataFrame()

df['Pais'] = countries
df['Confirmadors'] = confirmeds
df['Novos_Confirmados'] = new_cases
df['Obitos'] = deaths
df['Novos_Obitos'] = new_deaths
df['Recuperados'] = recov
df['Novos_Recuperados'] = new_recov
df['Casos_Ativos'] = act_cases
df['Total_Testes'] = total_tests
df['Populacao'] = population

df
```
